## Build a Model with Log-Transformed Features and Target

### Data Preparation

Choose up to 3 of the features you investigated, and set up an X dataframe containing the log-transformed versions of these features as well as a y series containing the log-transformed version of the target.

---

<details>
    <summary style="cursor: pointer"><b>Hint (click to reveal)</b></summary>

If you are planning log transform a feature measured in _years_ (e.g. `YearRemodAdd`) consider shifting the data first. For example, you might subtract 1900 or 1910 from the year, so that a 1% increase in year is closer to meaning 1 year rather than 20 years.

</details>

```python
# Your code here - prepare data for modeling
#create X variable as the log of dataframe containing three columns
X = np.log(ames[['GrLivArea','1stFlrSF','GarageArea']])
#define the y_variable as the log of the SalePrice column 
y= np.log(ames['SalePrice'])
    
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model
import statsmodels.api as sm

#define the model
model = sm.OLS(y, sm.add_constant(X))
#fit the model 
model_results = model.fit()
#print the results 
print(model_results.summary())
```

### Model Evaluation and Interpretation

How did the model perform? How might we interpret its coefficients? Create as many cells as needed.

```python
# Your code here - evaluate the model
"""
Constant (Intercept): const = 4.1063

    Interpretation: The intercept is the expected value of the log-transformed SalePrice when all the log-transformed predictors are equal to zero.
    Since all predictors are log-transformed, this intercept doesn't represent the price in actual dollars but rather the baseline log of the SalePrice 
    when the log-transformed predictors are zero. 
    In practice, this intercept may not always have a direct real-world interpretation, especially when dealing with log-transformed variables.
"""
```

```python
"""
Coefficient of GrLivArea: 0.5806

    Interpretation: A 1% increase in GrLivArea (the above-ground living area) is associated with approximately a 0.5806% increase in SalePrice,
    assuming all other variables are held constant.
    Explanation: In the context of log-log regression, the coefficients represent elasticities. Specifically, the coefficient 0.5806 means that for every 
    1% increase in GrLivArea, the SalePrice increases by approximately 0.5806%. This suggests that larger living areas are positively associated with higher
    sale prices, which is intuitive in the real estate market.
    """
```
